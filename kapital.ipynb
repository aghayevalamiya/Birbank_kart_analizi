```python
gelir_seviyyesi = []
for i in df['AVERAGE_SALARY'].iloc[:510]:
    if i < 600:
        gelir_seviyyesi.append('asagi')
    elif i > 600 and i < 1200:
        gelir_seviyyesi.append('orta')
    else:
        gelir_seviyyesi.append('yuxari')
```

```python
df['gelir_seviyyesi'] =gelir_seviyyesi
df['gelir_seviyyesi']
```

```python
df['contract_duration'] = (pd.to_datetime(df['CLOSE_DATE']).dt.year - pd.to_datetime(df['OPEN_DATE']).dt.year)
```
